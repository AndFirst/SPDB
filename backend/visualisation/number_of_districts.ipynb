```python
import json

import matplotlib.pyplot as plt

with open("../results/number_of_districts.json", "r") as f:
    data = json.load(f)


keys = data.keys()  
elapsed_time = []
optimized_time = []
default_time = []
optimized_distance = []
default_distance = []
optimized_yield = []
default_yield = []


for key in keys:
    entries = data[key]
    n = len(entries)
    
    
    sum_elapsed_time = 0
    sum_opt_time = 0
    sum_def_time = 0
    sum_opt_distance = 0
    sum_def_distance = 0
    sum_opt_yield = 0
    sum_def_yield = 0
    
    
    for entry in entries:
        sum_elapsed_time += entry["elapsed_time"]
        sum_opt_time += entry["optimized_stats"]["time"]
        sum_def_time += entry["default_stats"]["time"]
        sum_opt_distance += entry["optimized_stats"]["distance"]
        sum_def_distance += entry["default_stats"]["distance"]
        sum_opt_yield += entry["optimized_stats"]["num_yield_directions"]
        sum_def_yield += entry["default_stats"]["num_yield_directions"]
    
    
    elapsed_time.append(sum_elapsed_time / n)
    optimized_time.append(sum_opt_time / n)
    default_time.append(sum_def_time / n)
    optimized_distance.append(sum_opt_distance / n)
    default_distance.append(sum_def_distance / n)
    optimized_yield.append(sum_opt_yield / n)
    default_yield.append(sum_def_yield / n)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(keys, elapsed_time, marker='o', label="redni czas wykonania", color="#4BC0C0")

plt.xlabel("Liczba punktw")
plt.ylabel("Czas (sekundy)")
plt.title("Czas wykonania algorytmu w zalenoci od liczby punktw (z rnych dzielnic).")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/rozne_dzielnice_wydajnosc.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(keys, optimized_time, marker='o', label="Optymalizacja czasu z uwzgldnieniem pierwszestwa", color="#36A2EB")
plt.plot(keys, default_time, marker='o', label="Optymalizacja odlegoci", color="#FF6384")

plt.xlabel("Liczba punktw")
plt.ylabel("Czas (minuty)")
plt.title("Porwnanie czasw przejazdu w zalenoci od liczby punktw (z rnych dzielnic).")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/rozne_dzielnice_czas.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(keys, optimized_distance, marker='o', label="Optymalizacja czasu z uwzgldnieniem pierwszestwa", color="#36A2EB")
plt.plot(keys, default_distance, marker='o', label="Optymalizacja odlegoci", color="#FF6384")

plt.xlabel("Liczba punktw")
plt.ylabel("Dystans (km)")
plt.title("Porwnanie dugoci przejazdu w zalenoci od liczby punktw (z rnych dzielnic).")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/rozne_dzielnice_dystans.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(keys, optimized_yield, marker='o', label="Optymalizacja czasu z uwzgldnieniem pierwszestwa", color="#36A2EB")
plt.plot(keys, default_yield, marker='o', label="Optymalizacja odlegoci", color="#FF6384")

plt.xlabel("Liczba punktw")
plt.ylabel("Liczba kierunkw z pierwszestwem")
plt.title("Porwnanie liczby kierunkw z pierwszestwem w zalenoci od liczby punktw (z rnych dzielnic).")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/rozne_dzielnice_kierunki.png")
plt.show()
```
